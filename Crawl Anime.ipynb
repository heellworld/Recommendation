```python
#Thu Nhập các đường link ở trang web

# Khởi tạo trình duyệt với tùy chọn
options = webdriver.ChromeOptions()
# Thêm các tùy chọn khác nếu cần
driver = webdriver.Chrome(options=options)

# Mở trang web ban đầu
driver.get("https://animevietsub.fan/")
sleep(random.randint(5, 10))

# Get links
elems = driver.find_elements(By.CSS_SELECTOR, ".sub-menu [href]")
links = [elem.get_attribute('href') for elem in elems]

# Đóng trình duyệt sau khi hoàn thành
driver.quit()
```

```python
# Loại bỏ các đường link có dạng .html
links = [link for link in links if not link.endswith('.html')]

# Biểu thức chính quy để kiểm tra xem có số năm ở cuối không
year_pattern = re.compile(r'/\d{4}/$')

# Loại bỏ các đường link có số năm ở cuối
links = [link for link in links if not year_pattern.search(link)]

links
```

```python
# Đọc tệp CSV vào DataFrame
df_links = pd.read_csv('linksAnime.csv')
#Lấy cột đầu tiên và chuyển thành list
links = df_links.iloc[:, 0].tolist()
```

```python
links
```

```python
# Vào từng đường link để lấy thông tin phim và các đường link con

# Khởi tạo trình duyệt với tùy chọn
options = webdriver.ChromeOptions()
# Thêm các tùy chọn khác nếu cần
driver = webdriver.Chrome(options=options)

# Khởi tạo DataFrame trước vòng lặp
final_data = {"title": [], "view": [], "rating": [], "eps": [], "img": [], "linktitle": [], "listlink": []}

for link in links:
    driver.get(link)
    sleep(random.randint(5, 10))

    count = 1
    
    while True:
        try:
            # Kiểm tra xem trang cụ thể có tồn tại hay không
            if count == 1:
                next_page_film = link
            else:
                next_page_film = f"{link}trang-{count}.html"
            
            driver.get(next_page_film)
            print("Crawl Page " + str(count))
            
            #Get link của title
            elems_title_link = driver.find_elements(By.CSS_SELECTOR, ".TPostMv [href]")
            #Get title
            elems_title = driver.find_elements(By.CSS_SELECTOR, ".TPostMv .Title")
            #Get view của title
            elems_view = driver.find_elements(By.CSS_SELECTOR, ".TPostMv .Year")
            #Get rating của title
            elems_rating = driver.find_elements(By.CSS_SELECTOR, ".TPostMv .anime-extras")
            #Get eps của title
            elems_eps = driver.find_elements(By.CSS_SELECTOR, ".TPostMv .mli-eps")
            #Get img của title
            elems_img = driver.find_elements(By.CSS_SELECTOR, ".TPostMv .Image [src]")
            
            # Kiểm tra xem có phần tử nào không
            if not elems_title:
                print("No more elements found.")
                break

                        # Thêm dữ liệu vào DataFrame trực tiếp
            final_data["title"].extend([elem.text for elem in elems_title])
            final_data["view"].extend([elem.text for elem in elems_view])
            final_data["rating"].extend([elem.text for elem in elems_rating])
            final_data["eps"].extend([elem.text for elem in elems_eps])
            final_data["img"].extend([elem.get_attribute('src') for elem in elems_img])

            final_data["linktitle"].extend([elem.get_attribute('href') for elem in elems_title_link])
            final_data["listlink"].extend([link] * len(elems_title))


            print("Clicked on button next page!")
            sleep(random.randint(1, 3))
            count += 1
        except ElementNotInteractableException:
            print("Element Not Interactable Exception!")
            break

# Tạo DataFrame cuối cùng sau khi vòng lặp kết thúc
if final_data["title"]:
    # Loại bỏ các giá trị rỗng hoặc null từ final_data["title"]
    final_data["title"] = [elem for elem in final_data["title"] if elem.strip()]

    # Loại bỏ các đường link trùng lặp với các đường link trong links
    final_data["linktitle"] = list(set(final_data["linktitle"]) - set(links))
    
    # Kiểm tra độ dài trước khi tạo DataFrame
    lengths = [len(final_data[key]) for key in final_data]
    max_length = max(lengths)
    
    # Nếu độ dài không khớp, thêm NaN để độ dài bằng nhau
    final_data = {key: final_data[key] + [np.nan] * (max_length - len(final_data[key])) for key in final_data}

    df = pd.DataFrame(final_data)
    print(df)
else:
    print("Không có dữ liệu để tạo DataFrame.")

# Đóng trình duyệt sau khi hoàn thành
driver.quit()
```

```python
df.to_csv('Title_Anime.csv', index=False)
```

```python
# Đọc tệp CSV vào DataFrame
df_links = pd.read_csv('Title_links.csv')
#Lấy cột đầu tiên và chuyển thành list
linktitle_list = df_links.iloc[:, 0].tolist()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
from time import sleep
import random

options = webdriver.ChromeOptions()
# Thêm các tùy chọn khác nếu cần
driver = webdriver.Chrome(options=options)

# Khởi tạo danh sách dữ liệu
describe_data = []
information_data = []

for i in range(len(linktitle_list)):
    link = linktitle_list[i]
    driver.get(link)
    sleep(random.randint(3, 6))

    try:
        # Thu thập mô tả
        elems_describe = driver.find_elements(By.CSS_SELECTOR, ".Description")
        describe_text = " ".join([elem.text for elem in elems_describe]) if elems_describe else ""
        describe_data.append(describe_text)

        # Thu thập thông tin
        elems_info = driver.find_elements(By.CSS_SELECTOR, ".InfoList .AAIco-adjust")
        info_text = " ".join([elem.text for elem in elems_info]) if elems_info else ""
        information_data.append(info_text)

        print("Processed link:", link)
    except NoSuchElementException:
        describe_data.append("")
        information_data.append("")
    except Exception as e:
        print(f"Error processing link {i}: {e}")
        describe_data.append("")
        information_data.append("")

# Đóng trình duyệt sau khi hoàn thành
driver.quit()

# Tạo DataFrame cho describe và information
df_describe = pd.DataFrame({"describe": describe_data})
df_information = pd.DataFrame({"information": information_data})

# In DataFrame cuối cùng
print("DataFrame Describe:")
print(df_describe)
print("DataFrame Information:")
print(df_information)
```

```python
df_describe.to_csv('decribe.csv', index=False)
df_information.to_csv('information.csv', index=False)
```
